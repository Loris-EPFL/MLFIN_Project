```python
# process_compustat.py
import pandas as pd
import matplotlib.pyplot as plt
from preprocessing.preprocessing_compustat_bis import CompustatPreprocessor


# Initialize the preprocessor
preprocessor1 = CompustatPreprocessor(date_column='datadate', id_column='cusip')

# Step 1: Load the data
# Replace with your actual file path
file_path = "datasets/Predictors/CompFirmCharac.csv"
preprocessor1.load(file_path, nrows=500000)

# Step 2: Explore the data to understand its structure
preprocessor1.explore_data()

# Step 3: Visualize missing values
preprocessor1.plot_missing_values(top_n=100)



# Step 4: Convert dates to proper format
preprocessor1.convert_dates()

coverage = preprocessor1.check_column_coverage()
coverage.to_csv("coverage.csv")

recommended_columns = preprocessor1.recommend_columns_for_sentiment(min_coverage=50)
print(recommended_columns)

# Step 5: Select important financial columns
preprocessor1.select_important_columns()

preprocessor1.create_sentiment_indicators()


preprocessor1.global_sort_by_date()




# Step 6: Fix data types
preprocessor1.fix_data_types()

# Step 7: Handle duplicates
preprocessor1.handle_duplicates()

# Step 5b: Clean string 'nan' values
preprocessor1.clean_string_nan_values()

# Step 8b: Replace remaining NaNs in string columns with empty strings
preprocessor1.handle_missing_values(method='empty_strings')


# # Step 9: Create basic financial ratios
# preprocessor.create_financial_ratios()

# Step 10: Get the processed data
processed_data = preprocessor1.get_data()

# Print summary of the processed data
print("\n=== Processed Data Summary ===")
print(f"Shape: {processed_data.shape}")
print("\nSample of processed data:")
print(processed_data.head())

# Step 11: Export the processed data
output_path = 'datasets/PostProcessed/Predictors/compustat_processed_1.csv'
preprocessor1.export_data(output_path, format='csv')
    

```
