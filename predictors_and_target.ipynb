```python
import pandas as pd
from utils import aggregate_earnings_calls, transform_earnings_calls, prepare_for_merge_ec, fill_missing_earning_calls, \
    transform_tenkreports, aggregate_tenkreports, prepare_for_merge_10k, fill_missing_10k
```

```python
# Load the processed target data
target = pd.read_csv('./clean/merged_compustat_returns_cleaned_data.csv')
target.head()
```

```python
# Load earnings calls data
earnings_calls = pd.read_parquet('./clean/sentiment_earning_calls.parquet')
earnings_calls.head()
```

```python
earnings_calls = transform_earnings_calls(earnings_calls)
earnings_calls
```

```python
weighted_sentiment_ec = aggregate_earnings_calls(earnings_calls)
weighted_sentiment_ec
```

```python
target, weighted_sentiment_ec = prepare_for_merge_ec(target, weighted_sentiment_ec)
```

```python
target.head()
```

```python
merged_earning_calls = pd.merge_asof(
    target,
    weighted_sentiment_ec,
    left_on='MthCalDt',
    right_on='mostimportantdateutc',
    by='gvkey',
    direction='backward'
)
```

```python
merged_earning_calls
```

```python
merged_earning_calls = fill_missing_earning_calls(merged_earning_calls)
```

```python
merged_earning_calls.isna().sum().sum()
```

```python
merged_earning_calls
```

```python
tenkreports = pd.read_parquet('./clean/sentiment_tenkreports.parquet')
tenkreports
```

```python
tenkreports = transform_tenkreports(tenkreports)
```

```python
weighted_sentiment_10k = aggregate_tenkreports(tenkreports)
weighted_sentiment_10k
```

```python
merged_earning_calls, weighted_sentiment_10k = prepare_for_merge_10k(merged_earning_calls, weighted_sentiment_10k)
```

```python
merged_datasets = pd.merge_asof(
    merged_earning_calls,
    weighted_sentiment_10k,
    left_on='MthCalDt',
    right_on='report_date',
    by='cik',
    direction='backward'
)
```

```python
merged_datasets = fill_missing_10k(merged_datasets)
```

```python
merged_datasets
```

```python
merged_datasets.to_parquet('./clean/merged_datasets_with_sentiment.parquet')
```
